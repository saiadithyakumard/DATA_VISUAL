```python
data = pd.read_csv("HistoricalQuotes.csv")

# Clean up column names by removing leading/trailing spaces
data.columns = data.columns.str.strip()

# Remove dollar signs and convert 'Close/Last' to a numeric type
data['Close/Last'] = data['Close/Last'].replace('[\$,]', '', regex=True).astype(float)

# Calculate the mean, median, mode, and standard deviation of the 'Close/Last' prices
mean_close = np.mean(data['Close/Last'])
median_close = np.median(data['Close/Last'])
mode_close = data['Close/Last'].mode()[0]  # Taking the first mode in case of multiple modes
std_dev_close = np.std(data['Close/Last'])

# Print the results
print(f"Mean of 'Close/Last' prices: {mean_close}")
print(f"Median of 'Close/Last' prices: {median_close}")
print(f"Mode of 'Close/Last' prices: {mode_close}")
print(f"Standard Deviation of 'Close/Last' prices: {std_dev_close}")
 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("tiktok_performance.csv")

# Basic Operation: Calculate Median of Likes
if 'Likes' in df.columns:
    median_likes = df['Likes'].median()
    print("\nMedian of Likes:", median_likes)

# Advanced Operation: Calculate Engagement Rate (Likes per 1000 Views) and Display High Engagement Videos
if {'Likes', 'Views'}.issubset(df.columns):
    df['Engagement_Rate'] = (df['Likes'] / df['Views']) * 1000
    high_engagement = df[df['Engagement_Rate'] > 10]
    print("\nHigh Engagement Videos (Engagement Rate > 10):\n", high_engagement[['Views', 'Likes', 'Engagement_Rate']].head())

    # Visualization of Engagement Rate Distribution
    df['Engagement_Rate'].plot(kind='hist', bins=20, title="Distribution of Engagement Rates")
    plt.xlabel('Engagement Rate per 1000 Views')
    plt.show()



```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a synthetic dataset
data = {
    'Category': ['Dance', 'Comedy', 'Education', 'Dance', 'Comedy', 'Education'] * 10,
    'Views': np.random.randint(1000, 10000, 60),
    'Likes': np.random.randint(100, 2000, 60),
    'Comments': np.random.randint(10, 500, 60),
    'Shares': np.random.randint(5, 300, 60),
}
df = pd.DataFrame(data)

# 1. Boxplot to compare Likes across Categories
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Likes', data=df)
plt.title('Comparison of Likes by Category')
plt.ylabel('Likes')
plt.show()

# 2. Pie Chart for the composition of total Views by Category
views_summary = df.groupby('Category')['Views'].sum().reset_index()
plt.figure(figsize=(8, 8))
plt.pie(views_summary['Views'], labels=views_summary['Category'], autopct='%1.1f%%', startangle=90)
plt.title('Composition of Views by Category')
plt.axis('equal')  # Equal aspect ratio ensures pie chart is circular
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)                 # Sine function

# 1. Line Plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Sine Wave', color='blue')
plt.title('Line Plot of Sine Function')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid(True)
plt.show()

# 2. Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 10]

plt.figure(figsize=(10, 5))
plt.bar(categories, values, color='orange')
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# 3. Scatter Plot
x_scatter = np.random.rand(50) * 10  # 50 random points for x
y_scatter = np.random.rand(50) * 10  # 50 random points for y

plt.figure(figsize=(10, 5))
plt.scatter(x_scatter, y_scatter, color='green', alpha=0.5)
plt.title('Scatter Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)
plt.show()

# 4. Histogram
data = np.random.randn(1000)  # 1000 random numbers from a normal distribution

plt.figure(figsize=(10, 5))
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
